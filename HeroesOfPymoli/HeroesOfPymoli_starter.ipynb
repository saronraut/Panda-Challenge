```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#created a variable that contained the data for Items
items = purchase_data[["Item ID", "Item Name","Price"]]

#using the len func found the total number of unique items in the list
total_unique = len(items["Item ID"].unique())

#using mean function on the Price list, the average price for the items are identified
avg = (items["Price"].mean())

#using the sum to add all the prices in the list to get the total revenue
total_rev = items["Price"].sum()

# with in the data using the count function, total number of purchase was counted.
num_purchase = items["Item ID"].count()

#using pandas items data frame was created, the use of dictionary was used to create the df
items_df = pd.DataFrame({"Number or Unique Items": [total_unique], "Average Price":[avg],"Number of Purchase": [num_purchase]
                        , "Total Revenue": [total_rev]})
#using the example for "Kickstarter Clean" Classroom activity, the float was formatted. 
items_df["Average Price"] = items_df["Average Price"].map("${:,.2f}".format)
items_df["Total Revenue"] = items_df["Total Revenue"].map("${:,.2f}".format)

items_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#using loc to specify the targetted data

#female data
female_df = purchase_data.loc[purchase_data["Gender"]== "Female", :]
female_purchase_count = female_df["Item ID"].count()
female_mean = female_df["Price"].mean()
female_ttl_purch = female_df["Price"].sum()
#thought process was to group Gender by User then get their purchase sum and then get the average 
#used the example from 'exlporing pandas' from supplement to build this thought process
female_per = female_df.groupby(['SN']).sum()
female_per = female_per["Price"].mean()


#male data
male_df = purchase_data.loc[purchase_data["Gender"]== "Male", :]
male_purchase_count = male_df["Item ID"].count()
male_mean = male_df["Price"].mean()
male_ttl_purch = male_df["Price"].sum()
#thought process was to group Gender by User then get their purchase sum and then get the average 
#used the example from 'exlporing pandas' from supplement to build this thought process
male_per = male_df.groupby(['SN']).sum()
male_per = male_per["Price"].mean()


#non-disclosed data
other_df = purchase_data.loc[purchase_data["Gender"]== "Other / Non-Disclosed", :]
other_purchase_count = other_df["Item ID"].count()
other_mean = other_df["Price"].mean() 
other_ttl_purch = other_df["Price"].sum()
#thought process was to group Gender by User then get their purchase sum and then get the average 
#used the example from 'exlporing pandas' from supplement to build this thought process
other_per = other_df.groupby(['SN']).sum()
other_per = other_per["Price"].mean()

#creating df utilizing the dictionary 
Purchase_Analy = pd.DataFrame({"Gender" : ["Female","Male", "Other / Non-Disclosed"],
                               "Purchase Count": [female_purchase_count,male_purchase_count,other_purchase_count],
                              "Average Purchase Price": [female_mean,male_mean,other_mean],
                              "Total Purchase Value": [female_ttl_purch,male_ttl_purch,other_ttl_purch],
                               "Avg Total Purchase per Person": [female_per,male_per,other_per]
                             })
#format data frame to show $ and round to 2 decimal points. Also to set index 
Purchase_Analy["Average Purchase Price"] = Purchase_Analy["Average Purchase Price"].map("${:,.2f}".format)
Purchase_Analy["Total Purchase Value"] = Purchase_Analy["Total Purchase Value"].map("${:,.2f}".format)
Purchase_Analy["Avg Total Purchase per Person"] = Purchase_Analy["Avg Total Purchase per Person"].map("${:,.2f}".format)
Purchase_Analy = Purchase_Analy.set_index("Gender")

Purchase_Analy
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#created a bin to group the data into age group, labels were created to match the group
bins = [0,9,14,19,24,29,34,39,50]
age_label = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#insert the age label into the dataframe, where Age is used as the matching criteria, similar to Vlookup
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],bins,labels=age_label)

#created a seperate data to make it easier to work with also, so to drop the duplicate values from the list. 
#this time used the groupby function to return the values for all the age group instead of manual calculation
group_df = purchase_data[["Age Group", "Age", "Gender", "SN"]].drop_duplicates()
group_df = group_df.groupby("Age Group")
age_count = group_df["Age"].count()

#player count was calculated in the begining of the task and has been saved as the variable (player_count)
percnt_count = age_count/player_count * 100

#create data frame using the collected information
#was not able to remove the column name (Age Group) to match the example. but all the information are same.
Age_df = pd.DataFrame({"Total Count" : age_count, "Percentage of Players": percnt_count })
Age_df["Percentage of Players"] = Age_df["Percentage of Players"].map("{:,.2f}%".format)
#print 
Age_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#created a variable to work with: to group the data
pop_item = purchase_data.groupby(["Item ID", "Item Name"])
item_purchased = pop_item["Item ID"].count()
sum_item = pop_item["Price"].sum()
#took the sum of the item and divide to get the original price of the item
price_item = sum_item/item_purchased

#created the Data frame based on the infromation from the top
Popular_item = pd.DataFrame({"Purchase Count": item_purchased, "Item Price": price_item,
                              "Total Purchase Value": sum_item})
Popular_item["Item Price"] =Popular_item["Item Price"].map("${:,.2f}".format)
Popular_item["Total Purchase Value"] = Popular_item["Total Purchase Value"].map("${:,.2f}".format)
#set the sort to descending for Purchase count
Popular_item = Popular_item.sort_values("Purchase Count", ascending = False)
Popular_item.head()
```
